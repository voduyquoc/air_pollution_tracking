```python
def get_SO2_subindex(x):
    if x < 20:
        return 1
    elif x < 80:
        return 2
    elif x < 250:
        return 3
    elif x < 350:
        return 4
    else:
        return 5
```

```python
def get_NO2_subindex(x):
    if x <= 20:
        return 1
    elif x <= 40:
        return 2
    elif x <= 100:
        return 3
    elif x <= 200:
        return 4
    else:
        return 5
```

```python
def get_PM10_subindex(x):
    if x <= 20:
        return 1
    elif x <= 35:
        return 2
    elif x <= 50:
        return 3
    elif x <= 100:
        return 4
    else:
        return 5
```

```python
def get_PM2_5_subindex(x):
    if x <= 10:
        return 1
    elif x <= 20:
        return 2
    elif x <= 25:
        return 3
    elif x <= 50:
        return 4
    else:
        return 5
```

```python
def get_O3_subindex(x):
    if x <= 60:
        return 1
    elif x <= 120:
        return 2
    elif x <= 180:
        return 3
    elif x <= 240:
        return 4
    else:
        return 5
```

```python
def get_CO_subindex(x):
    if x < 4400:
        return 1
    elif x < 9400:
        return 2
    elif x < 12400:
        return 3
    elif x < 15400:
        return 4
    else:
        return 5
```

```python
df['latitude'] = df['latitude'].astype(str)
df['longitude'] = df['longitude'].astype(str)
df['StationId'] = df['latitude'] + " & " + df['longitude']
```

```python
df["PM10_24hr_avg"] = df.groupby("StationId")["coarse_particulate_matter_pm10"].rolling(window = 24).mean().values
df["PM2.5_24hr_avg"] = df.groupby("StationId")["fine_particles_matter_pm2_5"].rolling(window = 24).mean().values
```

```python
df["NO2_SubIndex"] = df["nitrogen_dioxide_no2"].apply(lambda x: get_NO2_subindex(x))
```

```python
df["PM10_SubIndex"] = df["PM10_24hr_avg"].apply(lambda x: get_PM10_subindex(x))
```

```python
df["PM2_5_SubIndex"] = df["PM2.5_24hr_avg"].apply(lambda x: get_PM2_5_subindex(x))
```

```python
df["PMO3_SubIndex"] = df["ozone_o3"].apply(lambda x: get_O3_subindex(x))
```

```python
df.head()
```

```python
df["AQI_calculated"] = df[["NO2_SubIndex",
                           "PM10_SubIndex",
                           "PM2_5_SubIndex",
                           "PMO3_SubIndex"]].max(axis = 1)
```

```python
df["check"] = df["aqi"]-df["AQI_calculated"]
```
